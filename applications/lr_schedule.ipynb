## Cosine Anneling with Restarts

The implementation formula is taken from [AdamW paper](https://arxiv.org/pdf/1711.05101.pdf). Compare the plot with _SuppFigure 1_ from the paper.

```python
from loop.schedule import CosineAnnealingSchedule
```

```python
cos = CosineAnnealingSchedule()
```

```python
def run_schedule(sched, n):
    return zip(*[(i, sched.update()) for i in range(n)])
```

```python
def plot_schedule(sched, n=1000):
    xs, ys = run_schedule(sched, n)
    f, ax = plt.subplots(1, 1, figsize=(10, 8))
    ax.plot(xs, ys)
```

```python
plot_schedule(cos, n=1499)
```
